```python
# factorial function
def fac(n):
    if(n == 0):
        return 1
    return n * fac(n-1)
```

```python
# takes an x value to approximate and a list;
# appends the cummulative sum of the first 30 terms of the series
# to the given list
def taylorfy(x, f_taylor):
    for n in range(1,31):
        f_taylor.append(f_taylor[n-1] + (x ** n / fac(n)))
```

```python
# plots the cummulative sum of the Taylor series passed in as a list
# in blue; plots the number approxmiated in red; takes the range of the
# y axis as two arguments
def plot(x, f_taylor, min_y, max_y):
    value = [math.exp(x)] * 31
    plt.title('Taylor Approximation of e^({})'.format(str(x)))
    plt.xlabel('n')
    plt.ylabel('f_n(x)')
    plt.plot(range(31), f_taylor, 'bo', range(31), value, 'r:')
    plt.axis([0, 31, min_y, max_y]) 
    plt.show()
```

```python
# 3A
f_taylor = [[1], [1], [1]]

taylorfy(1, f_taylor[0])
plot(1, f_taylor[0], 0, math.ceil(math.exp(1)))
```

The Taylor series quickly approaches the actual value ($\mathrm{e}$). Since $T(0)=1$, the the proceeding terms need not be large for the sum to approach the value.

```python
taylorfy(20, f_taylor[1])
plot(20, f_taylor[1], 0, (math.ceil(math.exp(20)) + 5e7))
```

The terms of the Taylor series from $10 < n < 30$ are large, but approach 0 as $n \rightarrow \infty$. The approximations for $n< 30$ are therefore inaccurate.

```python
taylorfy(-20, f_taylor[2])
plot(-20, f_taylor[2], -2.5e7, 2.5e7)
```

The terms of the Taylor series from $10 < n < 30$ are both large and alternate signs, making the Taylor approximations for these $n$ inaccurate. The terms of the expansion for $n>30$ approach 0, and as $n \rightarrow \infty$, the Taylor approximation approaches the value $\mathrm{e}^{-20}$.
